```python
#DFT
import numpy as np
import matplotlib.pyplot as plt

def dft(arr):
    N=len(arr)
    X=np.zeros(N,dtype=complex)
    W=np.exp((-1j*2*np.pi)/N)
    for k in np.arange(N):
        for n in np.arange(N):
            X[k]+=arr[n]*W**(n*k)
    return np.round(X)

arr=[1,2,3,4]
print(dft(arr))
print(np.absolute(dft(arr)))
print(np.angle(dft(arr),deg=True))
fig,axs=plt.subplots(2,figsize=(10,10))
axs[0].stem(np.absolute(dft(arr)), use_line_collection = True)
axs[0].set_title("Magnitude")
axs[0].grid(True)
axs[1].stem(np.angle(dft(arr),deg=True), use_line_collection = True)
axs[1].set_title("Phase")
axs[1].grid(True)
# plt.show()
plt.savefig('DFT.pdf')
```
